### 修改预测图

```python
import os
import glob
import cv2
import numpy as np

# =========================
# 1. 路径设置
# =========================
input_folder = r"work_dirs/Crackseg0309output/group3/selected"      # 原始图片文件夹
output_folder = r"work_dirs/Crackseg0309output/group3/selected2"    # 输出图片文件夹
target_points_path = "points_scaled.npy"
```

```python
input_folder = r"work_dirs/Crackseg0309output/group3/a"      # 原始图片文件夹
output_folder = r"work_dirs/Crackseg0309output/group3/b"    # 输出图片文件夹
target_points_path = "points_scaled.npy"
os.makedirs(output_folder, exist_ok=True)

# =========================
# 2. 参数设置
# =========================
reverse_section_order = False
flip_y = False                  # 你前面说上下反了，所以这里设为 False
border_color = (70, 70, 70)     # 深灰色边框（BGR）
border_thickness = 10
black_thresh = 5                # <=5 都视为黑色，改成 0 就是只替换纯黑
supported_exts = ["*.png", "*.jpg", "*.jpeg", "*.bmp", "*.tif", "*.tiff"]

# =========================
# 3. 工具函数
# =========================
def ensure_closed(pts: np.ndarray) -> np.ndarray:
    if len(pts) < 3:
        raise ValueError("轮廓点数量太少，无法构成多边形。")
    if not np.allclose(pts[0], pts[-1]):
        pts = np.vstack([pts, pts[0]])
    return pts

def load_section_points(full_h: int) -> np.ndarray:
    target_points = np.load(target_points_path)
    target_points = np.squeeze(target_points)

    if target_points.ndim != 2:
        raise ValueError(f"target_points squeeze 后维度不是 2，当前 shape={target_points.shape}")

    if target_points.shape[1] > 2:
        target_points = target_points[:, :2]

    section_pts = target_points.astype(np.float32)
    section_pts = ensure_closed(section_pts)

    # 是否进行上下翻转
    if flip_y:
        section_pts[:, 1] = full_h - section_pts[:, 1]

    if reverse_section_order:
        section_pts = section_pts[::-1]

    return section_pts

def process_one_image(img_path: str, output_path: str):
    img = cv2.imread(img_path)
    if img is None:
        print(f"[跳过] 无法读取图片: {img_path}")
        return

    h, w = img.shape[:2]

    # 加载轮廓点
    section_pts = load_section_points(h)

    # 转成 int，并限制到图像范围内
    poly = np.round(section_pts).astype(np.int32)
    poly[:, 0] = np.clip(poly[:, 0], 0, w - 1)
    poly[:, 1] = np.clip(poly[:, 1], 0, h - 1)
    poly_cv = poly.reshape((-1, 1, 2))

    # 白底图
    result = np.full_like(img, 255)

    # 掌子面区域 mask
    mask = np.zeros((h, w), dtype=np.uint8)
    cv2.fillPoly(mask, [poly_cv], 255)

    # 只保留掌子面区域原图
    result[mask == 255] = img[mask == 255]

    # 将图中纯黑/近黑区域改成纯白
    black_mask = np.all(result <= black_thresh, axis=2)
    result[black_mask] = [128, 128, 128]

    # 画深灰色边框
    cv2.polylines(
        result,
        [poly_cv],
        isClosed=True,
        color=border_color,
        thickness=border_thickness,
        lineType=cv2.LINE_AA
    )

    cv2.imwrite(output_path, result)
    print(f"[完成] {img_path} -> {output_path}")

# =========================
# 4. 批量处理
# =========================
image_paths = []
for ext in supported_exts:
    image_paths.extend(glob.glob(os.path.join(input_folder, ext)))

image_paths = sorted(image_paths)

if len(image_paths) == 0:
    raise FileNotFoundError(f"在文件夹 {input_folder} 中没有找到图片。")

for img_path in image_paths:
    filename = os.path.basename(img_path)
    output_path = os.path.join(output_folder, filename)
    process_one_image(img_path, output_path)

print("全部处理完成。")
```
